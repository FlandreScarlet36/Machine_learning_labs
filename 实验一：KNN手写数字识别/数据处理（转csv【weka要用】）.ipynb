```python
import torch
import numpy as np
import pandas as pd
```

```python
# 导入数据
def Img2Mat(fileName):
    f = open(fileName)
    ss = f.readlines()
    l = len(ss)
    f.close()
    returnMat = np.zeros((l,256))
    returnClassVector = np.zeros((l,1))
    for i in range(l):
        s1 = ss[i].split()
        for j in range(256):
            returnMat[i][j] = np.float64(s1[j])
        clCount = 0
        for j in range(256,266):
            if s1[j] != '1':
                clCount += 1
            else:
                break
        returnClassVector[i] = clCount
    return returnMat,returnClassVector
```

```python
X,y = Img2Mat('semeion.data')
```

```python
shape = (1593,10)
zeros_tensor = torch.zeros(shape)
for index, item in enumerate(y):
    zeros_tensor[index][int(item)] = 1.0
semeion = pd.DataFrame(X)
semeion['257'] = zeros_tensor.numpy().tolist()
semeion.to_csv('weka.csv',index=False)
```
